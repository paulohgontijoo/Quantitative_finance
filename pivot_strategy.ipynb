```python
def normalize_prices(market_close):
    test = market_close.copy()
    test = test.dropna()
    last = test.iloc[-1]
    for i in range(len(test)):
        test.iloc[i] = test.iloc[i]/last
    return test
```

```python
def get_real_returns(market_close):
    returns = pd.DataFrame(index = market_close.iloc[0].index, columns=['returns'])
    for i in market_close.iloc[0].index:
        actual_price = market_close[i].iloc[-1]
        if(math.isnan(actual_price)):
            actual_price = market_close[i].dropna()
            actual_price = actual_price.iloc[-1]
        last_price = market_close[i].iloc[0]
        if(math.isnan(last_price)):
            last_price = market_close[i].dropna()
            last_price = last_price[0]
        ret_real = (actual_price-last_price)/last_price 
        returns.loc[i]['returns'] = ret_real
    return returns
```

```python
# Sharpe
def sharpe(ret_ativ, ret_free_risk, ativ_volatility):
    s = (ret_ativ - ret_free_risk)/ativ_volatility
    return s

def lenght_equalize(n,m):
    maior = n
    menor = m
    if maior >= menor:
        return menor
    else:
        return maior

def beta(stock, ibov):
    stock = stock.pct_change().dropna()
    ibov_ret = ibov['Adj Close'].pct_change().dropna()
    n = lenght_equalize(len(stock),len(ibov_ret))
    stock = stock[:n]
    ibov_ret = ibov_ret[:n]
    return cov(stock, ibov_ret)[0][1]/variance(ibov_ret)

def traynor(ret_ativ, ret_free_risk, beta):
    excess_ret = (float(ret_ativ) - ret_free_risk)
    return float(excess_ret/beta)

def alpha(ret_ativ, ret_free_risk, beta, ibov_ret):
    alpha = (ret_ativ - ret_free_risk - beta)*(ibov_ret-ret_free_risk)
    return alpha

def r_squared(ativ_ret, ibov_ret):
    n = lenght_equalize(len(ativ_ret),len(ibov_ret))
    ativ_ret = ativ_ret[:n]
    ibov_ret = ibov_ret[:n]
    correlation_matrix = np.corrcoef(ativ_ret, ibov_ret)
    correlation_xy = correlation_matrix[0,1]
    return correlation_xy**2
```

```python
import random

# Markowits
multiplicador = 25
n_days = len(stocks_all)
stocks = stocks_all['Adj Close']
log_ret = np.log(stocks / stocks.shift(1))

# Admite que a mesma sequencia randomica
np.random.seed(42)
num_ports = multiplicador * 500 * len(stocks.columns)
all_weights = np.zeros((num_ports, len(stocks.columns)))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)

for x in range(num_ports):
    # Weights
    weights = np.array(np.random.random(len(stocks.columns)))
    weights = weights/np.sum(weights)

    # Save weights
    all_weights[x,:] = weights

    # Expected return
    ret_arr[x] = np.sum( (log_ret.mean() * weights * n_days))

    # Expected volatility
    vol_arr[x] = np.sqrt(np.dot(weights.T, np.dot(log_ret.cov() * n_days, weights)))

    # Sharpe Ratio
    sharpe_arr[x] = ret_arr[x]/vol_arr[x]

# Para operar na venda utilize o agrmin()

maximo = sharpe_arr.argmax()
# maximo = sharpe_arr.argmin()

# maximo = vol_arr.argmax()
# maximo = vol_arr.argmin()

         
max_sr_ret = ret_arr[maximo]
max_sr_vol = vol_arr[maximo]
```
