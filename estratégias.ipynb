```python
def normalize_prices(market_close):
    test = market_close.copy()
    test = test.dropna()
    last = test.iloc[-1]
    for i in range(len(test)):
        test.iloc[i] = test.iloc[i]/last
    return test
```

```python
def get_real_returns(market_close):
    returns = pd.DataFrame(index = market_close.iloc[0].index, columns=['returns'])
    for i in market_close.iloc[0].index:
        actual_price = market_close[i].iloc[-1]
        if(math.isnan(actual_price)):
            actual_price = market_close[i].iloc[-2]
        last_price = market_close[i].iloc[0]
        if(math.isnan(last_price)):
            last_price = market_close[i].dropna()
            last_price = last_price[0]
        ret_real = (actual_price-last_price)/last_price 
        returns.loc[i]['returns'] = ret_real
    return returns
```

```python
import random

# Markowits
print("Insira um multiplicador para a modelagem: ")
multiplicador = int(input())
n_days = len(stocks_all)
stocks = stocks_all['Adj Close']
log_ret = np.log(stocks / stocks.shift(1))

# Admite que a mesma sequencia randomica
np.random.seed(42)
num_ports = multiplicador * 500 * len(stocks.columns)
all_weights = np.zeros((num_ports, len(stocks.columns)))
ret_arr = np.zeros(num_ports)
vol_arr = np.zeros(num_ports)
sharpe_arr = np.zeros(num_ports)

for x in range(num_ports):
    # Weights
    weights = np.array(np.random.random(len(stocks.columns)))
    weights = weights/np.sum(weights)

    # Save weights
    all_weights[x,:] = weights

    # Expected return
    ret_arr[x] = np.sum( (log_ret.mean() * weights * n_days))

    # Expected volatility
    vol_arr[x] = np.sqrt(np.dot(weights.T, np.dot(log_ret.cov() * n_days, weights)))

    # Sharpe Ratio
    sharpe_arr[x] = ret_arr[x]/vol_arr[x]

# Para operar na venda utilize o agrmin()

maximo = sharpe_arr.argmax()
# minimo = sharpe_arr.argmin()

# maximo = vol_arr.argmax()
# minimo = vol_arr.argmin()

         
max_sr_ret = ret_arr[maximo]
max_sr_vol = vol_arr[maximo]
```
